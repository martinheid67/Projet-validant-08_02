```python
import pandas as pd
import numpy as np 

data = pd.read_csv('openaq.csv', sep=';')
data
```

```python
data.dtypes
```

```python
data['Last Updated']=pd.to_datetime(data['Last Updated'], utc=True).dt.tz_convert(None)
data_f_dates = data.loc[(data['Last Updated'] >= '2021-08-06 00:00:00') & (data['Last Updated'] < '2021-08-07 00:00:00')]
data_f_dates
```

```python
data_f_dates.City
```

```python
data_f_dates_and_cities = data_f_dates.dropna(subset = ['City'])
data_f_dates_and_cities
```

```python
data_f_dates_and_cities.Value
```

```python
data_f_dates_and_cities_and_values = data_f_dates_and_cities[data_f_dates_and_cities['Value']>0]
data_f_dates_and_cities_and_values
```

```python
data_f_dates_and_cities_and_values.to_csv('exo2.csv')
```
