```python
import pandas as pd
import numpy as np 

data = pd.read_csv('Energie.csv', sep='|')
data
```

```python
data['SITE_NAME'] = data['SITE_NAME'].str[11:].str[:-1]
data
```

```python
data.drop('INVOICE_NAME', axis=1, inplace=True)
data
```

```python
data.BEGIN = pd.to_datetime(data.BEGIN)
data.END = pd.to_datetime(data.END)
data['PERIODE_FACTURATION'] = (data.END - data.BEGIN)
data
```

```python
data['PERIODE_FACTURATION'] = data['PERIODE_FACTURATION'].dt.days
data
```

```python
data.dtypes
```

```python
data['KWH_PERIOD'] = (data.KWH_PERIOD.replace(' ', '', regex=True)).apply(str).astype(int)
data
```

```python
data['DAILY_CONSUM'] = (data.KWH_PERIOD / data.PERIODE_FACTURATION).round(decimals = 2)
data
```

```python
data['NB_INVOICES_SITE'] = data['SITE_NAME'].map(data['SITE_NAME'].value_counts())
data
```

```python
data.to_csv('exo1.csv')
```
